```python
import os
import pandas as pd
import numpy as np
```

```python
def calc_thrm(df):
    P = df['POWER']
    FF = df['FUEL_FLOW']
    FL = df['FUEL_LHV']
    THRM_EFF = (P/(FF*FL))*100
    return THRM_EFF
```

```python
engine = pd.read_csv("D:/engine_metadata-fad08922e295794745ba1cd70379140f.csv")
site = pd.read_csv("D:/site_metadata-7fc535965d0c074cea0be6786fc9518e.csv")
```

```python
path = 'D:/all_data_v2-e22b278886ea35866986a8802e38ff75/all_data_v2'

# Change the directory
os.chdir(path)
# os.listdir()
DF = []
for filename in os.listdir():
#     print(os.path.join(path, filename))
    df=pd.read_csv(os.path.join(path, filename))
    # add column file_id
    df['FILE_ID'] = filename
    # merge 3 datasets
    df1 = df.merge(engine, how = 'left', on = 'FILE_ID')
    df2 = df1.merge(site, how = 'left', on = 'PLANT_NAME')
    # calculate thermal efficiency
    df2['THRM_EFF'] = calc_thrm(df2)
    DF.append(df2)
```

```python
Df_all = pd.concat(DF)
```

```python
# 
Df_all.rename(columns={'Unnamed: 0':'Date/Time'}, inplace = True)
# Formatted Datetime
Df_all["Date/Time"] = pd.to_datetime(Df_all["Date/Time"], format="%Y-%m-%d %H:%M:%S")
```

## Power for each plant

```python
unique_plant = Df_all.PLANT_NAME.unique()
# Power for each plant
for i in range(len(unique_plant)):
    df1 = Df_all[Df_all['PLANT_NAME']==unique_plant[i]]
    df2 = df1[['Date/Time','POWER']].groupby(df1['Date/Time'].dt.date).mean()
    plt.figure(figsize = (15, 5))
    sns.lineplot(data = df2, x ='Date/Time', y = 'POWER')
```

```python

df3 = Df_all[['Date/Time','POWER']].groupby(Df_all['Date/Time'].dt.date).mean()
plt.figure(figsize = (15, 5))
sns.lineplot(data = df3, x ='Date/Time', y = 'POWER')
```

## Find out where is the location of the 'Angelic-Reindeer' plant

```python

df_select = Df_all[Df_all['PLANT_NAME']=='ANGELIC-REINDEER']
gdf = ps.GeoDataFrame(df_select, 
                      geometry=ps.points_from_xy(df_select.LONGITUDE, df_select.LATITUDE))
# get build in dataset from geopandas
world_data = ps.read_file(ps.datasets.get_path('naturalearth_lowres'))
# Plot world map
axis = world_data.plot(color = 'lightblue')
gdf.plot(ax=axis, color = 'red')

fig = plt.gcf()
fig.set_size_inches(20,18)
plt.show()
```

It's in Europe!
